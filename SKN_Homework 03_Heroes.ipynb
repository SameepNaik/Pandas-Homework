```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head(10)
```

```python
# PLAYER COUNT, players who have made purchases

# Remove duplicate players from purchase list
pd_total_players_df = purchase_data.drop_duplicates(subset=["SN"])
#players = purchase_data["SN"].value_counts()
#players_df = pd.DataFrame(players)
total_players = len(pd_total_players_df)
total_players_df = pd.DataFrame([{"Total Players": total_players}])
#total_players_df_rename = total_players_df.rename(columns={"Total Players":})
total_players_df
```

```python
# PURCHASING ANALYSIS (TOTAL)

# Find number of uniqe items
unique_items = purchase_data["Item ID"].unique()
num_unique_items = len(unique_items)

# Find average price
#average_price = "%.2f" %((purchase_data["Price"].sum())/(len(purchase_data["Price"])))
average_price = purchase_data["Price"].mean()
average_price = "%.2f" %average_price

# Find number of puchases
num_purchases = purchase_data["Price"].count()

# Find total revenue
total_revenue = purchase_data["Price"].sum()

# Create data frame
purchasing_analysis_total_df = pd.DataFrame({"Number of Unique Items": [(num_unique_items)],
                                      "Average Price": ["$"+str(average_price)],
                                      "Number of Purchases": [num_purchases],
                                      "Total Revenue": ["$"+str(total_revenue)],})
purchasing_analysis_total_df
```

```python
# GENDER DEMOGRAPHICS

# Find the number of males
male_df = purchase_data.loc[purchase_data["Gender"] == "Male",:]
male_count = len(male_df["SN"].unique())
male_count

# Find the number of females
female_df = purchase_data.loc[purchase_data["Gender"] == "Female",:]
female_count = len(female_df["SN"].unique())
female_count

# Find the number of other / non-disclosed
other_df = purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed",:]
other_count = len(other_df["SN"].unique())
other_count

# Calculate gender percentages
total_count = male_count+female_count+other_count
male_per = "%.2f" %((male_count/total_count)*100)
female_per = "%.2f" %((female_count/total_count)*100)
other_per = "%.2f" %((other_count/total_count)*100)

# Create data frame
gender_df = pd.DataFrame({"": ["Male", "Female", "Other/Non=Disclosed"],
                          "Total Count": [male_count, female_count, other_count],
                         "Percetage of Players": [male_per, female_per, other_per]})
gender_df


#gender_count = purchase_data["Gender"].value_counts()
#gender_count
#gender_df = purchase_data.groupby(['Gender'])
#print(gender_df)
#gender_df.head()
#male = purchase_data["SN"].unique()
```

```python
# PURCHASING ANALYSIS (GENDER)

# Calculate female purchase count
female_purchase_count = female_df["Purchase ID"].count()

# Calculate female average purchase price
female_ave_price = female_df["Price"].mean()
female_ave_price = "%.2f" %female_ave_price

# Calculate female total purchase price
female_total_purchase = female_df["Price"].sum()

# Calculate average total purchase per female
female_total_ave_purchase = "%.2f" %((female_total_purchase/female_count))

# Calculate male purchase count
male_purchase_count = male_df["Purchase ID"].count()

# Calculate male average purchase price
male_ave_price = male_df["Price"].mean()
male_ave_price = "%.2f" %male_ave_price

# Calculate male total purchase price
male_total_purchase = male_df["Price"].sum()

# Calculate average total purchase per male
male_total_ave_purchase = "%.2f" %((male_total_purchase/male_count))

# Calculate other purchase count
other_purchase_count = other_df["Purchase ID"].count()

# Calculate other average purchase price
other_ave_price = other_df["Price"].mean()
other_ave_price = "%.2f" %other_ave_price

# Calculate other total purchase price
other_total_purchase = other_df["Price"].sum()

# Calculate average total purchase per other
other_total_ave_purchase = "%.2f" %((other_total_purchase/other_count))

# Create data frame
purchasing_analysis_gender_df = pd.DataFrame({
    "Gender": ["Female", "Male", "Other / Non-Disclosed"],
    "Purchase Count": [female_purchase_count, male_purchase_count, other_purchase_count],
    "Average Purchase Price": ["$"+female_ave_price, "$"+male_ave_price,"$"+other_ave_price],
    "Total Purchase Value": ["$"+str(female_total_purchase), "$"+str(male_total_purchase), 
                             "$"+str(other_total_purchase)],
    "Avg Total Purchase per Person": ["$"+female_total_ave_purchase, "$"+male_total_ave_purchase, 
                                      "$"+other_total_ave_purchase],})
purchasing_analysis_gender_df
```
